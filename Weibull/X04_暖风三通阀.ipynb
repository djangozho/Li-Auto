```python
#new_rate=0
new_rate = len(failure_data.loc[failure_data['use_days'] <= 30]) / len(risk_data)

out = pd.DataFrame()
out['use_days'] = [30 * (i + 1) for i in range(60)]  # 60个月，到1800天

def calculate_metrics(x):
    sf = output.distribution.SF(x)
    cdf_upper = output.distribution.CDF(CI_x=[x], CI=0.95)[-1]
    cdf_lower = output.distribution.CDF(CI_x=[x], CI=0.95)[0]
    # 计算总故障率：前30天故障率 + 存活到30天后的产品在后续时间的故障率
    avg = (new_rate + (1 - new_rate) * (1 - sf)) * 100
    upper = (new_rate + (1 - new_rate) * cdf_upper) * 100
    lower = (new_rate + (1 - new_rate) * cdf_lower) * 100
    return pd.Series([avg, upper, lower])

out[['平均(%)', '95%上限(%)', '95%下限(%)']] = out['use_days'].apply(calculate_metrics)

out.to_clipboard(index=False)

usedays = 1800
use = out.loc[out['use_days'] == usedays]

# 提取平均、上限、下限三个值
rate_ = use[['平均(%)', '95%上限(%)', '95%下限(%)']].values.flatten()
num_ = len(risk_data) * rate_ / 100

print(f"{usedays}天故障率在{min(rate_):.4f}%到{max(rate_):.4f}%之间，估计损失件数在{min(num_):.0f}到{max(num_):.0f}之间。更高天数数据缺乏证据，仅作为参考。")
```
